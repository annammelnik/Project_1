```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path1 = "movies_clean.csv"
path2 = "the_oscar_award (version 1).csv"
moviesdf = pd.read_csv(path1)
oscardf = pd.read_csv(path2)
moviesdf = moviesdf.rename(columns={"name":"film", "year":"year_film"})
merge = pd.merge(moviesdf, oscardf, on=["film", "year_film"])
merge
```

```python
just_actors = merge.loc[(merge['year_film'] >=2010)&((merge['category'] == 'ACTOR IN A LEADING ROLE') | (merge['category'] == 'ACTRESS IN A LEADING ROLE')), :]
just_actors
```

```python
plt.bar(just_actors['name'], just_actors['gross'])
plt.xticks(rotation = 'vertical')
plt.ylabel('Revenue in Billions (USD)')
plt.xlabel("Best Actor/Actress since 2010")
plt.title("Oscar Winning Actors vs Revenue")
plt.savefig('../PNG Files/oscar_winning_actors.png', format = 'png', bbox_inches='tight')
plt.show()
```

```python
oscar_path = 'the_oscar_award(version 2).csv'
oscar_nominees = pd.read_csv(oscar_path)
new_merge = pd.merge(moviesdf, oscar_nominees , on=["film", "year_film"])

best_pictures = new_merge.loc[(new_merge['year_film'] ==2017)&((new_merge['category'] == 'BEST PICTURE')), :].sort_values('gross', ascending = False).head(5)
best_pictures
explode = [0,0,.1,0,0]
best_pictures.set_index('film').plot.pie(y='gross', legend=False, explode = explode, startangle=90, shadow=True,autopct='%1.2f%%')
plt.axis('equal')
plt.ylabel('Percentage of Revenue')
plt.title("2018 Oscars Top 5 Grossing Best Picture Nominees")
plt.savefig('../PNG Files/2018_Oscars.png', format = 'png', bbox_inches='tight')
plt.show()
```
