```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
import numpy as np
from scipy.stats import linregress
```

```python
path1 = "movies_adjusted.csv"
path2 = "the_oscar_award(version 2).csv"

moviesdf = pd.read_csv(path1)
oscardf = pd.read_csv(path2)

moviesdf.head()
```

```python
oscardf.head()
```

```python
#changed path2


moviesdf = moviesdf.rename(columns={"title":"film", "year":"year_film"})
merge = pd.merge(moviesdf, oscardf, how="outer", on=["film", "year_film"])
merge=merge.fillna(0)
#merge.head()
#you wind up with multiple listings for movies with multiple noms
merge.head(50)
```

```python
#narrow down to films that made money since 2010
merge= merge.loc[(merge["gross"] >0) &(merge['year_film'] >=2010),:]
#merge.head()

#same as Christopher's

#added conditional of winner is true
just_actors = merge.loc[(merge['year_film'] >=2010)&(merge['winner'] == True)&((merge['category'] == 'ACTOR IN A LEADING ROLE') | (merge['category'] == 'ACTRESS IN A LEADING ROLE')), :]
#just_actors['gross'] =
```

```python
plt.bar(just_actors['name'], just_actors['gross'])
plt.xticks(rotation = 'vertical')
plt.ylabel('Revenue in Billions (USD)')
plt.xlabel("Best Actor/Actress since 2010")
plt.title("Oscar Winning Actors vs Revenue")
plt.show()
```

```python
#narrow down to coumns to graph, grouping by film in order to eliminiate duplicates
film_ct = merge[['film','winner_ct',"nominees", 'gross']].groupby("film").mean()

film_ct.head(10)
```

```python
noms=film_ct.loc[film_ct['nominees'] >0,:]
multinoms=film_ct.loc[film_ct['nominees'] >1,:]
winners=film_ct.loc[film_ct['winner_ct'] >0,:]
multiwin=film_ct.loc[film_ct['winner_ct'] >1,:]
losers=film_ct.loc[film_ct['nominees'] == 0,:]
```

```python
#First set variables for plot, calling on filtered groups above
datasets=[winners['gross'], multiwin['gross'], noms['gross'], multinoms['gross'],losers['gross']]
labels = ['One Winner','Multiple Winners', 'One Nominee', 'Multiple Nominees', 'No Nominees']
#red_diamond = dict(markerfacecolor='r', marker='D'), flierprops=red_diamond

#Plot it

fig1, ax1 = plt.subplots()
ax1.boxplot(datasets, labels=labels, patch_artist=True, layout=(2, 1))
           

#label it
ax1.set_title('Gross By Oscars')
ax1.set_ylabel('Gross')
#ax1.yaxis.grid(True)
plt.show()
```
