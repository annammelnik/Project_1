```python
import pandas as pd
```

```python
#source: https://www.kaggle.com/rounakbanik/the-movies-dataset?select=movies_metadata.csv


movies_df = pd.read_csv('archive/movies_metadata.csv', low_memory=False)
#movies_df.dropna(inplace = True) - did not do this here because it immediatly dropped it to 600 entries
movies_df.head()
#count-45460
```

```python
#take necessary columns
#add only english language films?
#did we decide on years?
movies_core=movies_df[['title', 'budget', 'revenue', 'id']]
#movies_core=movies_df[['title', 'budget', 'revenue', 'id', 'original_language', 'release_date']]
movies_core.head()
```

```python
#change budget to numbers
movies_core['budget']=pd.to_numeric(movies_core['budget'], errors='coerce')
```

```python
#change id to numbers
movies_core['id']=pd.to_numeric(movies_core['id'], errors='coerce')
```

```python
movies_core.drop(movies_core[movies_core['budget'] < 10000].index, inplace = True)
#count8590
```

```python
movies_core.drop(movies_core[movies_core['revenue'] < 10000].index, inplace = True)
#count5280
```

```python
#open new csv to grab IMBD numbers 
links_df = pd.read_csv('archive/links.csv', low_memory=False)
links_df.head()
```

```python
#Rename column name to merge on, drop the one we dont need
links_df =links_df.drop(columns={"tmdbId"})
links_df =links_df.rename(columns={"movieId":'id'})
links_df.head()
```

```python
#merge links into core list to grab IMDB numbers on id number
core_w_imdb=movies_core.merge(links_df, how='left', on="id")
core_w_imdb
```

```python
#drop movies w/o IMDB info 
movies_clean=core_w_imdb.dropna()
movies_clean.count()
```

```python
movies_clean.to_csv("clean_mov_metadata_mr.csv", encoding="utf-8", index=False)
```
