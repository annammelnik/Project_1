```python
import pandas as pd
import numpy as np

# load files
path1 = "movie_budget.csv"
path2 = "inflation.csv"
path3 = "the_oscar_award (version 1).csv"
```

```python
movie_data = pd.read_csv(path1)
movie_data
```

```python
inflation_mult = pd.read_csv(path2)
inflation_mult.head()
#clean up inflation data
inflation_clean = inflation_mult[["year", "Multiplier"]]
inflation_clean

#Add one to the percent to find the mult
inflation_clean["Multiplier"]=inflation_clean["Multiplier"] +1
inflation_clean
```

```python
#empty lists
titles = []

budgets = []
adjusted_budgets = []

domestics = []
adjusted_domestics = []

internationals = []
adjusted_internationals = []

worldwides = []
adjusted_worldwides = []

#budget
i = 0

for year in inflation_clean["year"]:
    year_movie = movie_data.loc[movie_data["year"] == year,:]
    for budget in year_movie['budget']:
        adjusted_budget=budget*inflation_clean.iloc[i,1]
        budgets.append(budget)
        adjusted_budgets.append(adjusted_budget)
    i+=1
#domestic
i = 0

for year in inflation_clean["year"]:
    year_movie = movie_data.loc[movie_data["year"] == year,:]
    for domestic in year_movie['domestic']:
        adjusted_domestic=domestic*inflation_clean.iloc[i,1]
        domestics.append(domestic)
        adjusted_domestics.append(adjusted_domestic)
    i+=1
#international
i=0

for year in inflation_clean["year"]:
    year_movie = movie_data.loc[movie_data["year"] == year,:]
    for international in year_movie['international']:
        adjusted_international=international*inflation_clean.iloc[i,1]
        internationals.append(international)
        adjusted_internationals.append(adjusted_international)
    i+=1

#worldwide
i=0 
for year in inflation_clean["year"]:
    year_movie = movie_data.loc[movie_data["year"] == year,:]
    for worldwide in year_movie['worldwide']:
        adjusted_worldwide=worldwide*inflation_clean.iloc[i,1]
        worldwides.append(worldwide)
        adjusted_worldwides.append(adjusted_worldwide)
    i+=1
i=0
for year in inflation_clean["year"]:
    year_movie = movie_data.loc[movie_data["year"] == year,:]
    for title in year_movie['title']:
        titles.append(title)
        
```

```python
adjusteddf = pd.DataFrame({"title":titles, 
                             "Adjusted Budget":adjusted_budgets, 
                             "Adjusted Domestic":adjusted_domestics, 
                             "Adjusted International":adjusted_internationals, 
                             "Adjusted Worldwide":adjusted_worldwides})
adjusteddf.head()
```

```python
movie_data = movie_data[['year','title','director','writer','producer','mpaa','main_actor_1','main_actor_2','main_actor_3','main_actor_4','genre_1','genre_2','genre_3','genre_4']]
merged = pd.merge(movie_data, adjusteddf, how='outer', on ='title')
merged_clean = merged.dropna(subset=['Adjusted Worldwide', 'Adjusted International', 'Adjusted International', 'Adjusted Budget'])
merged_clean
```
